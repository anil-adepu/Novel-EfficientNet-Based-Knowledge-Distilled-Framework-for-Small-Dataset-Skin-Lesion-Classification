```python
import numpy as np, pandas as pd, os
from keras.preprocessing import image
from PIL import Image
import os, random, cv2, copy
from os import path


def return_list(data_path, data_type):
    file_list = [file for file in os.listdir(data_path) if file.lower().endswith(data_type)]
    # print(str(len(file_list)))
    return file_list

def mk_dir(dir_path):
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
    return dir_path

data_type = '.jpg'
csv_path = 'D:/Project/'
FOLDS = 5

train= pd.read_csv(csv_path + 'HAM10000_metadata.csv')
cols= train.columns
classes = train['dx'].unique()
class_wise=[]
for i in  classes:
    class_wise.append(list(train.loc[train['dx']==i].image_id))

for x in class_wise:
    random.shuffle(x)

strat=[[] for i in range(FOLDS)]

for c, x in enumerate(class_wise):
    i=0
    for j in range(FOLDS):
        step = int(np.floor(len(x)/FOLDS))
        k= min(len(x), i + step)
        strat[j]+= x[i:k]
        i=k

    if c:
        for id, y in enumerate(x[i:len(x)]):
            strat[id].append(y)
    else:
        for id, y in enumerate(x[i:len(x)]):
            strat[len(strat)-1-id].append(y)

    print('----------------------------------------------------------------------------', len(x))

cnt=set()
for x in strat:
    for y in x:
        cnt.add(y)
print(len(cnt))

cnt=0
for x in strat:
        cnt += len(x)
print(cnt)
for x in strat:
    print(len(x), len(x)*100/32701)


train_path='D:/Project/CenterCropped384/train/'
files=return_list(train_path, '.jpg')

data_save_path = mk_dir('D:/Project/Stratified_Train')

for fold, fold_files in enumerate(strat):
    fold_wise_data_save_path = mk_dir('D:/Project/Stratified_Train/Fold' + str(fold) + '/')
    for file in fold_files:
        temp_txt = file + '.jpg'
        org_img = np.asarray(image.load_img(train_path + temp_txt))
        org_img = Image.fromarray(org_img)
        org_img.save(path.join((fold_wise_data_save_path) + file + '.jpg'))
```

```python
import numpy as np, pandas as pd, os
import matplotlib.pyplot as plt, cv2
import tensorflow as tf, re, math
from keras.preprocessing import image
from PIL import Image
from os import path

## INPUT DATASET: https://www.kaggle.com/datasets/aniladepu/isic2020-256x256-jpg
GCS_PATH = 'D:/Project/Stratified_Train/'
GCS_PATH2 = 'D:/Project/CenterCropped384/test/'
#IMGS2 = os.listdir(GCS_PATH2)
#print(len(IMGS2))

#print(GCS_PATH, GCS_PATH2)

def return_list(data_path, data_type):
    file_list = [file for file in os.listdir(data_path) if file.lower().endswith(data_type)]
    # print(str(len(file_list)))
    return file_list

def mk_dir(dir_path):
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
    return dir_path

def removeHair(image):
    grayScale = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

    kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(17,17))
    blackhat = cv2.morphologyEx(grayScale, cv2.MORPH_BLACKHAT, kernel)

    # apply thresholding to blackhat
    _,threshold = cv2.threshold(blackhat,36,255,cv2.THRESH_BINARY)

    final_image = cv2.inpaint(image,threshold,1,cv2.INPAINT_TELEA)
    return final_image


check = []
tmp = 'D:/Project/Stratified_Inpainted_384/train/'

for i in range(5):
    data_save_path = tmp + 'Fold'+str(i)
    data_save_path = mk_dir(data_save_path)
    files = return_list(GCS_PATH + 'Fold'+str(i), '.jpg')

    for lineIdx in range(len(files)):
        temp_txt = files[lineIdx]
        print('Processing Img {idx}: {temp_txt}'.format(idx=lineIdx + 1, temp_txt=temp_txt))
        # load image
        check.append(temp_txt)
        org_img = np.asarray(image.load_img(GCS_PATH + 'Fold'+str(i) + '/' + temp_txt))
        inpainted_img = removeHair(org_img)
        finalImg = Image.fromarray(inpainted_img)
        finalImg.save(path.join(data_save_path+'/', temp_txt))


print(len(check))


check = []
tmp = mk_dir('D:/Project/Stratified_Inpainted_384/test/')
data_save_path = mk_dir(tmp)

files = return_list(GCS_PATH2, '.jpg')
for lineIdx in range(len(files)):
    temp_txt = files[lineIdx]
    print('Processing Img {idx}: {temp_txt}'.format(idx=lineIdx + 1, temp_txt=temp_txt))
    # load image
    check.append(temp_txt)
    org_img = np.asarray(image.load_img(GCS_PATH2 + temp_txt))
    inpainted_img = removeHair(org_img)
    finalImg = Image.fromarray(inpainted_img)
    finalImg.save(path.join(data_save_path, temp_txt))

print(len(check))
```

```python
import numpy as np, pandas as pd, os
import matplotlib.pyplot as plt, cv2
import tensorflow as tf, re, math
from keras.preprocessing import image
from PIL import Image
from os import path

GCS_PATH = 'D:/Project/Stratified_Inpainted_384/train/' + 'Fold'
df = pd.read_csv('D:/Project/HAM10000_Train_Edited.csv')
test_df = pd.read_csv('D:/Project/ISIC_2019_Test_Metadata.csv')
```

```python
cats = ['sex','anatom_site_general'] 
for c in cats:
    df[c],mp = df[c].factorize()
    test_df[c],mp2 = test_df[c].factorize()
#     #print(mp,mp2)
print('Imputing Age NaN count =',df.age_approx.isnull().sum())
df.age_approx.fillna(df.age_approx.mean(),inplace=True)
df['age_approx'] = df.age_approx.astype('int')
print('Imputing Age NaN count =',test_df.age_approx.isnull().sum())
test_df.age_approx.fillna(test_df.age_approx.mean(),inplace=True)
test_df['age_approx'] = test_df.age_approx.astype('int')
```

```python
def _bytes_feature(value):
  """Returns a bytes_list from a string / byte."""
  if isinstance(value, type(tf.constant(0))):
    value = value.numpy() # BytesList won't unpack a string from an EagerTensor.
  return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
  """Returns a float_list from a float / double."""
  return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
  """Returns an int64_list from a bool / enum / int / uint."""
  return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
```

```python
def serialize_example(feature0, feature1, feature2, feature3, feature4, feature5, feature6):
  feature = {
      'image': _bytes_feature(feature0),
      'image_name': _bytes_feature(feature1),
      'age_approx': _int64_feature(feature2),
      'anatom_site_general': _int64_feature(feature3),
      'sex': _int64_feature(feature4),
      'target': _bytes_feature(feature5),
      'target_id': _int64_feature(feature6)
  }
  example_proto = tf.train.Example(features=tf.train.Features(feature=feature))
  return example_proto.SerializeToString()
```

```python
for j in range(5):
    PATH = GCS_PATH + str(j) + '/'
    print(); print('Writing TFRecord %i...'%(j))
    IMGS = os.listdir(PATH)
    with tf.io.TFRecordWriter('D:/Project/Stratified_Inpainted_384/tfrec/withoutstringencode/train%.2i-%i.tfrec'%(j,len(IMGS))) as writer:
        for k in range(len(IMGS)):
#             img = np.asarray(image.load_img(PATH+IMGS[k]))
            img = cv2.imread(PATH+IMGS[k])
#             img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Fix incorrect colors
            img = cv2.imencode('.jpg', img, (cv2.IMWRITE_JPEG_QUALITY, 94))[1].tostring()
            name = IMGS[k].split('.')[0]
            row = df.loc[df.image==name]
            example = serialize_example(
                img, 
                str.encode(name),
                row.age_approx.values[0],                        
                row.anatom_site_general.values[0],
                row.sex.values[0],
                str.encode(row.target.values[0]),                
                row.target_id.values[0])
            writer.write(example)
            if k%100==0: print(k,', ',end='')
```
